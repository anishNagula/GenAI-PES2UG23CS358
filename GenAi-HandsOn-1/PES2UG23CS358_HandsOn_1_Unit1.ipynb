### **Experiment 1: Text Generation**

```python
prompt = "The future of Artificial Intelligence is"
```

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

for name, model_id in models.items():
    print(f"\n{name} output:")
    try:
        generator = pipeline("text-generation", model=model_id)
        output = generator(prompt, max_length=30)
        print(output)
    except Exception as e:
        print("Error:", e)
```

### **Experiment 2: Masked Language Modeling (Missing Word):**

```python
sentences = {
    "BERT": "The goal of Generative AI is to [MASK] new content.",
    "RoBERTa": "The goal of Generative AI is to <mask> new content.",
    "BART": "The goal of Generative AI is to <mask> new content."
}

for name, model_id in models.items():
    print(f"\n{name} predictions:")
    try:
        fill = pipeline("fill-mask", model=model_id)
        outputs = fill(sentences[name])
        for o in outputs[:3]:
            print(o["token_str"], o["score"])
    except Exception as e:
        print("Error:", e)
```

### **Experiment 3: Question Answering**

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"
```

```python
for name, model_id in models.items():
    print(f"\n{name} answer:")
    try:
        qa = pipeline("question-answering", model=model_id)
        result = qa(question=question, context=context)
        print(result)
    except Exception as e:
        print("Error:", e)
```

| **Task** | **Model** | **Classification** | **Observation (What actually happened?)** | **Architectural Reason** |
| --- | --- | --- | --- | --- |
| **Generation** | BERT | Failure | Repeated punctuation without semantic continuation | Encoder-only model with no autoregressive decoding |
|  | RoBERTa | Failure | Echoed the prompt without generating new tokens | Optimized encoder-only architecture and no decoder |
|  | BART | Success | Generated new tokens but semantics were noisy | Encoder–decoder architecture supports generation |
| **Fill-Mask** | BERT | Success | Correctly predicted “create”, “generate”, “produce” with high confidence | Pretrained using Masked Language Modeling (MLM) |
|  | RoBERTa | Success | Produced correct and diverse verb predictions | Optimized MLM pretraining with larger data |
|  | BART | Partial | Produced weaker and lower-confidence predictions | Masking supported but not primary objective |
| **QA** | BERT | Partial | Extracted relevant risks but incomplete answer | Base model not fine-tuned for question answering |
|  | RoBERTa | Failure | Selected an irrelevant phrase from the context | Randomly initialized QA head without fine-tuning |
|  | BART | Partial | Returned the entire sentence as the answer | Seq2seq bias and lack of extractive QA training |
